## Preparo el DataSet para crear los algoritmos/modelos precisos

*  Como ya fue mencionado, al estar trabajando con un conjunto de datos desbalanceado (la mayoria no pertenecen al target), el algoritmo va a tender a aprender que el cliente, sea cual sea, bajo las caracteristicas que sean, no va a pertenecer al target. Es decir, no va a poder obtener credito.
*  Para que esto no suceda vamos a proceder a balancearlos usando el metodo de sobremuestreo. El metodo consiste en añadir datos de clientes que pertenecen al target (Target=1) para que la cantidad de personas que no pertenecen sea igual a la cantidad de personas que si lo hacen. Asi, el dataset estaria balanceado y hay menos probabilidad de que haya overfitting.
*  La desventaja de utilizar este metodo es que el overfitting puede ocurrir porque van a haber clientes con los mismos datos, por ende mismas caracteristicas, ya que no vamos a 'inventar' datos de un cliente, sino que volver a agregar datos de clientes existentes.


```python
#Importo las librerias de nuevo por comodidad
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Separo los datos por clase

```python
datos_no_target = df2[df2['Target']==0]
datos_target = df2[df2['Target']==1]
```

### Sobremuestreo con  remplazo

```python
sobremuestreo_datos_target = datos_target.sample(n=len(datos_no_target), replace=True, random_state=42)
sobremuestreo_datos_target
```

### Preparo los datos para aplicar a los modelos de clasificacion

```python
df4 = pd.concat([sobremuestreo_datos_target, datos_no_target])
df4
```

```python
#Observamos la dimension de nuestro nuevo DataSet
df4.shape
```
